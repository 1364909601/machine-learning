```python
#按照item_second_cate_cd分组，然后再按照out_wh_num_7d_full_band为第二分组，再按照out_qtty降序排列
```

```python
arr = np.array([])
row5_group = skuData.groupby('item_second_cate_cd')
for group_name, group_eles in row5_group:
    print(group_name)
    group_2 = group_eles.groupby(by='out_wh_num_7d_full_band',sort=True)
    for group2_name, group2_eles in group_2:        
        print(group2_name)
        group2_eles = group2_eles.sort_values(by='out_qtty',ascending=False)
        print(group2_eles)
        if len(arr)==0:
            arr=group2_eles.as_matrix()
        else:
            arr = np.concatenate((arr,group2_eles.as_matrix()) )
    
```

```python
#coding=utf-8  
import pandas as pd  
import numpy as np  
#以下实现排序功能。  
series=pd.Series([3,4,1,6],index=['b','a','d','c'])  
frame=pd.DataFrame([[2,4,1,5],[3,1,4,5],[5,1,4,2]],columns=['b','a','d','c'],index=['one','two','three'])  
print(frame)  
print(series)  

#series的排序API
print('series通过索引进行排序：')  
print(series.sort_index())  
print('series通过值进行排序:')  
print(series.sort_values())  

#dataframe的排序API

print('dataframe根据行索引进行降序排序（排序时默认升序，调节ascending参数）：')  
print(frame.sort_index(ascending=False))  
print('dataframe根据列索引进行排序：')  
print(frame.sort_index(axis=1))  
print('dataframe根据值进行排序：')  
print(frame.sort_values(by='a'))  
print('通过多个索引进行排序：')  
print(frame.sort_values(by=['a','c'])) 
```
